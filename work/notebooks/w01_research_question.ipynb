**Research question:** Which pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring, given limited review capacity?

**Who acts on it:** a content reviewer / SEO specialist who can only inspect a limited number of pages (top-K) per cycle.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd

path = "../../data/raw/content_refresh_anonymized.csv"
if not os.path.exists(path):
    path = "data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(path)
print(f"Rows: {len(df):,}")

down_share = (df["trend_direction"] == "down").mean()
print(f"Declining (trend == 'down'): {down_share:.1%}")

dd = ((df["trend_direction"] == "down") & (df["impressions_90d"] >= 100)).sum()
print(f"Declining with demand (impressions_90d >= 100): {dd:,}")

sv = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()
print(f"Stale visible (500+ impressions, 180+ days no update): {sv:,}")
```
